```python
# LIST OPERATIONS â€“ SINGLE PROGRAM (COMMON ONES)

# 1. Creating a list
lst = [10, 20, 30, 40]
print("Initial list:", lst)

# 2. Accessing elements
print("First element:", lst[0])
print("Last element:", lst[-1])
print("Slice [1:3]:", lst[1:3])

# 3. Modifying elements
lst[1] = 25
print("After modification:", lst)

# 4. Adding elements
lst.append(50)
print("After append:", lst)

lst.extend([60, 70])
print("After extend:", lst)

lst.insert(2, 15)
print("After insert:", lst)

# 5. Removing elements
lst.remove(25)
print("After remove:", lst)

lst.pop()
print("After pop:", lst)

lst.pop(2)
print("After pop at index 2:", lst)

# 6. Length
print("Length of list:", len(lst))

# 7. Searching
print("Index of 40:", lst.index(40))
print("Count of 10:", lst.count(10))
print("Is 40 in list?", 40 in lst)

# 8. Sorting
lst.sort()
print("After sort:", lst)

lst.sort(reverse=True)
print("After reverse sort:", lst)

# 9. Reversing
lst.reverse()
print("After reverse:", lst)

# 10. Copying
new_lst = lst.copy()
print("Copied list:", new_lst)

# 11. Mathematical operations
print("Maximum:", max(lst))
print("Minimum:", min(lst))
print("Sum:", sum(lst))

# 12. List concatenation & repetition
combined = lst + [100, 200]
print("After concatenation:", combined)

repeated = lst * 2
print("After repetition:", repeated)

# 13. Iteration
print("Iterating through list:")
for item in lst:
    print(item, end=" ")
print()

# 14. List comprehension
squares = [x * x for x in lst]
print("Squares using list comprehension:", squares)

# 15. Clearing list
lst.clear()
print("After clear:", lst)
```
