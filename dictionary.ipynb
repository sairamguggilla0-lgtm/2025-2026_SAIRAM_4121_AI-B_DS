```python
# 1. Creating dictionaries
d = {1: "Apple", 2: "Banana", 3: "Cherry"}
print("Initial dictionary:", d)

d2 = dict(name="John", age=25, city="Delhi")
print("Dictionary using dict():", d2)

# 2. Accessing values
print("Value for key 1:", d[1])
print("Value using get():", d.get(2))
```
