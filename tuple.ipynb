```python
# TUPLE OPERATIONS â€“ SINGLE PROGRAM (COMMON ONES)

# 1. Creating tuples
t = (10, 20, 30, 40, 20)
print("Initial tuple:", t)

t2 = tuple([1, 2, 3])     # from list
print("Tuple from list:", t2)

single = (5,)            # single-element tuple
print("Single element tuple:", single)

# 2. Accessing elements
print("First element:", t[0])
print("Last element:", t[-1])
print("Slice [1:4]:", t[1:4])

# 3. Length
print("Length of tuple:", len(t))

# 4. Searching
print("Index of 30:", t.index(30))
print("Count of 20:", t.count(20))
print("Is 40 in tuple?", 40 in t)

# 5. Iterating through tuple
print("Iterating through tuple:")
for item in t:
    print(item, end=" ")
print()

# 6. Tuple concatenation & repetition
t3 = t + t2
print("After concatenation:", t3)

t4 = t * 2
print("After repetition:", t4)

# 7. Mathematical operations
print("Maximum:", max(t))
print("Minimum:", min(t))
print("Sum:", sum(t))

# 8. Tuple packing & unpacking
packed = 1, 2, 3
print("Packed tuple:", packed)

a, b, c = packed
print("Unpacked values:", a, b, c)

# 9. Converting tuple to other types
lst = list(t)
print("Tuple to list:", lst)

st = set(t)
print("Tuple to set:", st)

# 10. Sorting tuple (returns list)
sorted_tuple = sorted(t)
print("Sorted tuple:", sorted_tuple)

# 11. Nested tuple
nested = ((1, 2), (3, 4))
print("Nested tuple:", nested)
print("Access nested element:", nested[1][0])
```
